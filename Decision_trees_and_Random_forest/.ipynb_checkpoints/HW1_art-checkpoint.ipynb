2. (1 балл) Для регрессионного дерева необходимо использовать такой критерий:
    $$H(R) = \min_c \frac{1}{|R|} \sum_{(x_i, y_i) \in R} (y_i - c)^2$$
    
    Докажите, что минимум H(R) достигается при $c$:

    $$ c = \frac{1}{|R|} \sum_{(x_j, y_j) \in R} y_j$$

$$\frac{1}{|R|} \sum_{(x_i,y_i) \in R} (y_i - c)^2 = f(c)$$
$$\frac{\partial f(c)}{\partial c} = -\frac{2}{|R|} \sum_{(x_i,y_i) \in R} (y_i - c)$$
$$-\frac{2}{|R|} \sum_{(x_i,y_i) \in R} (y_i - c) = 0 \Longrightarrow \sum_{(x_i,y_i) \in R} y_i  - |R|c = 0$$
$$\Longrightarrow$$
$$c = \frac{1}{|R|} \sum_{(x_i,y_i) \in R} y_i$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# generate example
x_shape = 300
x = np.arange(x_shape) / 100
y = x**3 * np.sin(x**3) + np.random.random(x_shape)
```

```python
plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
from sem_dt_rf.decision_tree import criterio
from sem_dt_rf.decision_tree import decision_tree

TreeRegressor = decision_tree.RegressionDecisionTree(max_depth = 20, min_leaf_size = 5)
TreeRegressor.fit(x[:,np.newaxis],y)

X_shape = 3000
X = np.arange(X_shape)[:, np.newaxis] / 1000
y_pred = TreeRegressor.predict(X)

plt.figure()
plt.scatter(x, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X, y_pred, color="cornflowerblue", label="prediction", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("DecisionTree Regression")
plt.legend()
plt.show()
```

```python
TreeRegressor = decision_tree.RegressionDecisionTree(max_depth = 10, min_leaf_size = 5)
TreeRegressor.fit(x[:,np.newaxis],y)

X_shape = 3000
X = np.arange(X_shape)[:, np.newaxis] / 1000
y_pred = TreeRegressor.predict(X)

plt.figure()
plt.scatter(x, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X, y_pred, color="cornflowerblue", label="prediction", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("DecisionTree Regression")
plt.legend()
plt.show()
```

```python
from sem_dt_rf.random_forest.random_forest import RandomForestClassifier as MyRandomForestClassifier
from sem_dt_rf.random_forest.sampler import BaseSampler, ObjectSampler, FeatureSampler
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import fetch_covtype
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import balanced_accuracy_score
from sklearn.model_selection import train_test_split
import lightgbm as lgb
import xgboost as xgb
```

```python
cov_type = fetch_covtype()
```

```python
print(cov_type.data)
```

```python
print(cov_type.target)
```

```python
unique_elements = np.unique(cov_type.target)
n_classes = len(unique_elements)
n_classes
```

```python
x_train, x_test, y_train, y_test = train_test_split(cov_type.data, cov_type.target - 1)
```

```python
MyRF = MyRandomForestClassifier(
    n_estimators=100,
    max_objects_samples=0.9,
    max_features_samples=0.8,    
    max_depth=8,
    min_samples_leaf=1
)

Skclf = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=0)

params_lgb = {
    'num_leaves': 2 ** 8,
    'min_data_in_leaf': 50,
    'objective': 'multiclass',
    'num_classes': n_classes,
    'max_depth': 8,
    "boosting": "rf",
    "seed": 0,
    "bagging_frequency": 0.65,
    "subsample": .632,
    "subsample_freq": 1,
    "verbose": -1,
}
data_lgb = lgb.Dataset(data=x_train, label=y_train)

data_xgb = xgb.DMatrix(x_train, label=y_train)

params_xgb = {
    "booster": "gbtree",
    "subsample": .632,
    "max_depth": 8,
    "num_parallel_tree": 100,
    "objective": "multi:softmax",
    "num_class": n_classes,
    "eta":1,
    "random_state": 0,
    "n_jobs": -1,
}
```

# MyRandomForestClassifier

```python
%%time
MyRF.fit(x_train, y_train)
```

```python
%%time
y_pred = MyRF.predict(x_test)
```

```python
balanced_accuracy_score(y_test, y_pred)
```

#  lightgbm

```python
%%time
clf = lgb.train(params_lgb, train_set=data_lgb, num_boost_round=100)
```

```python
%%time
y_pred = clf.predict(x_test).argmax(axis=1)
```

# xgboost

```python
%%time
bst = xgb.train(params_xgb, data_xgb, num_boost_round=1)
```

```python
%%time
y_pred = bst.predict(xgb.DMatrix(x_test))
```

```python
balanced_accuracy_score(y_test, y_pred)
```
